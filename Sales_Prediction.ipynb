### Análisis de los Valores Faltantes
En el Dataframe:

*   Item_Weight: Tiene 1463 valores faltantes.
*   Outlet_Size: Tiene 2410 valores faltantes.

```python
df["Item_Weight"].describe()
```

```python
df.groupby("Item_Type")["Item_Weight"].describe()
```

> Se propone imputar los datos faltantes de "Item_Weight" basado en la media de cada "Item_Type" por las siguientes razones:

> *   La media y la desviación estándar varían significativamente entre los diferentes tipos de artículos, lo que sugiere que los pesos dentro de cada tipo son más similares entre sí que con otros tipos. Imputar la media dentro de cada tipo ayuda a mantener esta coherencia.
*   La imputación basada en la media por tipo de artículo asegura que los valores faltantes sean rellenados de manera representativa según la distribución de cada categoría, minimizando el sesgo.


---

> Para "Outlet_Size" se creará una nueva categoría llamada "Unknown" para los datos faltantes, ya que existirá mucho sesgo asumir el 28% de los datos con otro método.



### Abordar valores faltantes

```python
promedio_tipo = df.groupby("Item_Type")["Item_Weight"]
df["Item_Weight"] = promedio_tipo.transform(lambda x: x.fillna(x.mean()))
```

```python
df.groupby("Item_Type")["Item_Weight"].describe()
```

```python
df["Outlet_Size"] = df["Outlet_Size"].fillna("Unknown")
```
